```python
sentences = []
words = []
filtered = []
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
for i in data.Title:
    i = i.translate(str.maketrans('', '', string.punctuation))
    a = sent_tokenize(i)
    sentences.append(a)
    b = word_tokenize(i)
    words.append(b)
    c = [w.lower() for w in b if not w.lower() in stop_words]
    filtered.append(c)
```

```python

data['sentence_tokenized'] = sentences
data['word_tokenized'] = words
data['without_stop_words'] = filtered
data.head()
```
