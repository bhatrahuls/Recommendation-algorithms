```python
import pandas
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import string
import time
stemmer = PorterStemmer()
```

```python
data=pandas.read_csv("C:/Users/Rahul Bhat/Desktop/data.csv")
data.head()
```

```python
sentences = []
words = []
filtered = []
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
for i in data.Sentence:
        i = i.translate(str.maketrans('', '', string.punctuation))
        a = sent_tokenize(i)
        sentences.append(a)
        b = word_tokenize(i)
        words.append(b)
        c = [w.lower() for w in b if not w.lower() in stop_words]
        filtered.append(c)
data.head()
```

```python
data['sentence_tokenized'] = sentences
data['word_tokenized'] = words
data['filtered'] = filtered
data.head()
```

```python
data['stemmed_words']=data['filtered'].apply(lambda f: [stemmer.stem(word) for word in f])
data.head()
```

```python
inv_index = {}
permuterm_index={}

for pos,term in enumerate(data['stemmed_words']):
 for  word in term:
    if word in inv_index:
      if pos not in inv_index[word]:
        inv_index[word].append(pos)
    else:
      inv_index[word]=[]
      inv_index[word].append(pos)
      permuterm=set()
      w=word+"$"
      for i in range(len(word)):
        w=w[1 : ] + w[0 : 1]
        permuterm.add(w)
      permuterm_index[word]=[permuterm]


for i in inv_index:
  print(i,":",inv_index[i], end='\n\n')
  print(i,":",permuterm_index[i], end='\n\n')
```
